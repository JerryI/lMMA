```c++
unsigned constexpr const_hash(char const *input) {
    return *input ?
        static_cast<unsigned int>(*input) + 33 * const_hash(input + 1) :
        5381;
}
```

```c++
unsigned const_hash(string c) {
    return const_hash(c.c_str());
}
```
